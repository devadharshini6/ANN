```python
forest1= forest.iloc[:,2:30]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(forest1)
forest_norm = sc.transform(forest1)
forest_norm
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 28)
pca_values = pca.fit_transform(forest_norm)
pca_values
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
final = pd.concat([pd.DataFrame(pca_values[:,0:24],columns=['pc1','pc2','pc3','pc4','pc5','pc6','pc7',
 'pc8','pc9','pc10','pc11','pc12','pc13','pc14',
 'pc15','pc16','pc17','pc18','pc19','pc20','pc21',
 'pc22','pc23','pc24']),
 forest[['size_category']]], axis = 1)
final.size_category.replace(('large','small'),(1,0),inplace=True)
final
```

```python
import seaborn as sns
fig=plt.figure(figsize=(20,15))
sns.scatterplot(data=final)
```

```python
array = final.values
X = array[:,0:24]
Y = array[:,24]
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(10, input_dim=24, activation='relu'))
model.add(Dense(6,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, Y, validation_split=0.3, epochs=150, batch_size=10)
```

```python
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

the obtained accuracy for the model is 93%
